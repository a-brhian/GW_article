Neste exercício, vamos analisar a temperatura de cocção de dois tipos de alimentos: carne bovina e frango. O objetivo é verificar se há uma diferença significativa entre as temperaturas de cocção desses dois alimentos.


```r
set.seed(123)

# Temperaturas para carne bovina (média 70, desvio padrão 5)
carne_bovina <- round(rnorm(30, mean = 70, sd = 5),1)

# Temperaturas para frango (média 75, desvio padrão 6)
frango <- round(rnorm(30, mean = 75, sd = 6),1)

# Organizar os dados em um data frame
dados <- data.frame(
  Temperatura = c(carne_bovina, frango),
  Tipo_Alimento = rep(c("Carne Bovina", "Frango"), each = 30)
)

# Exibir os primeiros dados
View(dados)
```

#Diagrama de Caule e Folha *(stem-and-leaf)*

Considere que desejaríamos verificar as informações dos a temperatura de cocção da carne e do frango por meio do Diagrama de Caule e Folha.

A função que utilizaremos para isso é `stem()`.

```r
# Gráfico de caule e folha para carne bovina
carne_bovina
cat("Gráfico de caule e folha para carne bovina:\n")
stem(carne_bovina)
```

```r
# Gráfico de caule e folha para frango
sort(frango)
stem(frango)
stem(frango,scale=3)
```

#Histograma

Calculando o número de intervalos pelo método da *raiz quadrada*.

```r
num_bins <- 3
num_bins
```

O histograma no R é feito por meio da função `hist()`

```r
hist(frango, breaks = num_bins, main = "Histograma com Raiz Quadrada de Intervalos",
     xlab = "Valor", ylab = "Frequência", col = "skyblue", border = "black")
```

A função `hist()` do R estima o tamanho do *intervalo* por meio da Regra de Sturges:

$$
k = \lceil \log_2(n) + 1 \rceil
$$

```r
num_bins <- ceiling(log2(length(frango))+1)
num_bins
```

```r
hist(frango, main = "Histograma com Raiz Quadrada de Intervalos",
     xlab = "Temperatura de Cocção", ylab = "Frequência", col = "orange", border = "red")
```

#Medidas de localização

##Média Amostral

A média é calculada por meio da função `mean()`.

```r
mean(frango)
```

##Mediana

A mediana é calculada por meio da função `median` ou `quantile()`.

```r
median(frango)
cat("\n")
quantile(frango,probs=0.5)
```

##Quartis

Os quartis são calculados por meio da função `quantile()`.

```r
cat("1° quartil\n")
quantile(frango,probs = 0.25)
cat("\n2° quartil ou mediana\n")
quantile(frango,probs = 0.5)
cat("\n3° quartil\n")
quantile(frango,probs = 0.75)
```

Assim, todas essas medidas de localização poderiam ser representadas dentro do histograma:

```r
hist(frango, main = "Histograma com Raiz Quadrada de Intervalos",
     xlab = "Valor", ylab = "Frequência", col = "skyblue", border = "black")

legend("topright", legend = c("Média", "Quartis"),
       col = c("red", "orange"), lty = c(2, 1), lwd = 5, cex = 2, bty = "n")

abline(v=mean(frango),col=adjustcolor("red",.4),lty=2,lwd=5)
abline(v=quantile(frango,probs = c(.25,.5,.75)),col=adjustcolor("orange",.4),lty=1,lwd=5)
```

#Medidas de dispersão

##Desvio padrão

O desvio padrão amostral é calculado por meio da função `sd()`.

```r
sd(frango)
```

A variância amostral é dada pelo quadrado do desvio padrão amostral ou é calculado pela função `var()`.

```r
var(frango)
cat("\n")
(sd(frango))^2
```

##Desvio interquartílico

O desvio interquartílico é feito pela diferença entre o 3º e o 2º quartil.

```r
desvio.int <- quantile(frango,probs=.75) - quantile(frango,probs=.25)
desvio.int
```
